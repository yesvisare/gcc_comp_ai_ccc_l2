## Section 2: GDPR Analysis (7 Principles + 8 Rights)

### GDPR Overview

**7 Core Principles (Article 5):**
1. Lawfulness, Fairness, Transparency
2. Purpose Limitation
3. Data Minimization
4. Accuracy
5. Storage Limitation
6. Integrity and Confidentiality
7. Accountability

**8 Data Subject Rights:**
1. Right to Access (Article 15)
2. Right to Rectification (Article 16)
3. Right to Erasure (Article 17)
4. Right to Restriction of Processing (Article 18)
5. Right to Data Portability (Article 20)
6. Right to Object (Article 21)
7. Rights Related to Automated Decision-Making (Article 22)
8. Right to Lodge a Complaint

**Penalty:** â‚¬20M or 4% of global revenue

## Section 3: SOC 2 Analysis (5 Trust Service Criteria)

### SOC 2 Overview

**5 Trust Service Criteria:**
1. **Security (required):** Access control, MFA, audit logging
2. **Availability:** Uptime SLAs, disaster recovery, redundancy
3. **Processing Integrity:** Input validation, error handling
4. **Confidentiality:** Encryption at rest and in transit
5. **Privacy:** Consent management, privacy notices

**Type I vs Type II:**
- **Type I:** Point-in-time assessment (design only)
- **Type II:** 6-12 months operational evidence (design + effectiveness)

**Key Requirement:** Enterprise contracts typically require SOC 2 Type II certification

````python
# Analyze SOC 2 compliance
soc2_analyzer = SOC2Analyzer()
soc2_result = soc2_analyzer.analyze(sample_rag)

print(f"SOC 2 Compliance Score: {soc2_result.compliance_score:.2%}")
print(f"Controls Checked: {soc2_result.total_controls_checked}")
print(f"Compliant: {soc2_result.compliant_controls}")
print(f"Non-Compliant: {soc2_result.non_compliant_controls}")
print(f"\nRemediation Hours: {soc2_result.total_remediation_hours}")

# Note: SOC 2 is certification-based (no penalty risk like GDPR/HIPAA)

# Expected: 70-90% compliance for typical RAG architecture
````

## Section 5: HIPAA Analysis (26 Safeguards)

### HIPAA Security Rule Overview

**26 Safeguards:**

**Administrative (12):**
- Security management process
- Workforce security
- Security awareness training
- Business Associate Agreements (BAA)

**Physical (6):**
- Facility access controls
- Workstation security
- Device and media controls

**Technical (8):**
- Access control (unique user IDs)
- Audit controls (7-year retention)
- Encryption and decryption
- Transmission security

**Penalty:** $50,000 per violation
**Critical:** BAA signatures required from ALL vendors processing PHI

````python
# Add HIPAA-specific fields to RAG architecture
hipaa_rag = sample_rag.copy()
hipaa_rag["training"] = {"hipaa_training": True}
hipaa_rag["vendors"] = [
    {"name": "OpenAI", "baa_signed": True},
    {"name": "Pinecone", "baa_signed": True}
]
hipaa_rag["access_control"]["unique_user_ids"] = True
hipaa_rag["infrastructure"] = {"facility_access_control": True}

# Analyze HIPAA compliance
hipaa_analyzer = HIPAAAnalyzer()
hipaa_result = hipaa_analyzer.analyze(hipaa_rag)

print(f"HIPAA Compliance Score: {hipaa_result.compliance_score:.2%}")
print(f"Controls Checked: {hipaa_result.total_controls_checked}")
print(f"Compliant: {hipaa_result.compliant_controls}")
print(f"Non-Compliant: {hipaa_result.non_compliant_controls}")
print(f"\nTotal Penalty Risk: ${hipaa_result.total_penalty_risk:,}")

# Expected: 80-100% compliance with proper BAAs and training
````

## Section 10: GCC Multi-Tenant Scenarios

### Real-World GCC Challenge

**Scenario:** GCC serving 50+ business units across:
- **25 in US:** HIPAA + SOC 2 + CCPA
- **20 in EU:** GDPR + SOC 2 + ISO 27001
- **10 in India:** DPDPA + ISO 27001 + SOC 2

### Per-Tenant Chargeback Model

```
Cost = Base (â‚¹3K) + Framework Premium

Framework Premium:
- GDPR: â‚¹1,000
- SOC 2: â‚¹1,500
- ISO 27001: â‚¹1,200
- HIPAA: â‚¹2,000
```

````python
# Calculate per-tenant costs for different profiles
base_cost = 3000  # INR
framework_premium = {
    "GDPR": 1000,
    "SOC2": 1500,
    "ISO27001": 1200,
    "HIPAA": 2000
}

tenant_profiles = [
    {"name": "US Healthcare BU", "frameworks": ["HIPAA", "SOC2"]},
    {"name": "EU Finance BU", "frameworks": ["GDPR", "SOC2", "ISO27001"]},
    {"name": "India IT Services BU", "frameworks": ["ISO27001", "SOC2"]}
]

print("Per-Tenant Monthly Cost Breakdown")
print("=" * 70)

for profile in tenant_profiles:
    cost = base_cost + sum(framework_premium.get(f, 0) for f in profile["frameworks"])
    print(f"\n{profile['name']}:")
    print(f"  Frameworks: {', '.join(profile['frameworks'])}")
    print(f"  Monthly Cost: â‚¹{cost:,}")
    print(f"  Annual Cost: â‚¹{cost * 12:,}")

# Expected: Cost breakdown per tenant profile
````
